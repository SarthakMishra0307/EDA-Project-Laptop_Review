```python
labels = []
avg_prices = []
for gb, _df in df.groupby("graphic_card_gb"):
    labels.append(f"{gb}GB")
    avg_prices.append(_df["latest_price"].mean())

fig = mtb.figure()  
  
fig.set_figheight(4)  
fig.set_figwidth(15)  

mtb.subplot(1,2,1)
mtb.title("Prices of Different Graphic Card(Sizes)" )
mtb.barh(labels, avg_prices, color=["grey","wheat","orange","red","hotpink"])

values = df["graphic_card_gb"].value_counts()

mtb.subplot(1,2,2)
mtb.title("Size of graphic cards common in the dataset")
mtb.pie(values.values, labels=values.index ,autopct="%.1f%%")
mtb.legend()
mtb.show()
```
